```
import numpy as np
from matplotlib import pyplot as plt
```

```
#Esta es la difusión de calor en el espacio 2D.
#Se utiliza el método explícito de diferencias finitas.
# longitud en la dirección x
L = 1.0
# longitud en la dirección y
W = L
# Tiempo final
Tend = 80
maxk = 300
dt = Tend/maxk
n = 20
u = np.zeros((n+1,n+1,maxk+1))
x = np.zeros(n+1)
y = np.zeros(n+1)
time=np.zeros(maxk+1)
# k = np.zeros((n+1,n+1,maxk+1))
```

**CONDICIÓN INCIAL Y PARTE DELA CONDICIÓN DEL ENTORNO**

```
u[n][n][maxk] = 70
dx = L/n
dy = W/n
# dx = dy = h
h = dx
b = dt/(h*h)
cond = .002  # conductividad termica
spheat = 1.0 # calor especifico
rho = 1 # densidad
a = cond/(spheat*rho)
alpha = a*b
```

```
for i in range (1,n+1):
    x[i]=(i-1)*h; # use dx = dy = h 
    y[i]=(i-1)*h;
#condición de contorno
for k in range (1,maxk+1):
    time[k]= (k-1)*dt;

    for j in range (1,n+1):
        u[1][j][k]= 300*(k<120+70);
#calculando el metodo de diferencias finitas
for k in range (1,maxk):
     for j in range (2,n):
        for i in range (2,n):
            u[i][j][k+1]=0.*dt/(spheat*rho)
            +(1-4*alpha)*u[i][j][k]
            +alpha*u[i-1][j][k]+u[i+1][j][k]
            +u[i][j-1][k]+u[i][j+1][k]

          #VISUALIZACIÓN
        ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```
